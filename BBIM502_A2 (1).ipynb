```python
import os
import datetime

try:
    from RoomModule import  Room
    from CustomerModule import Customer
    from RoomAllocationModule import RoomAllocation
except ImportError as ex:
    print(ex)
    print("Please check your Modules")

list0fRooms = []
no0fRooms = 0
filePath = ""
filePathBackUp = ""
list0fRoomAllocations = []

# Main function
def main():
    # Initialising integer variable for number of room and
    # file path for both save allocations and back up file
    global no0fRooms, filePath, filePathBackUp
    # Initialising and declaring path for
    folder_Path = os.path.expanduser("~/Documents")
    # Create specific file for both save allocations and back up
    filePath = os.path.join(folder_Path, "lhms_764702333.txt")
    filePathBackUp = os.path.join(folder_Path, "lhms_764702333_backup.txt")
    menu()


# Method for main menu
def menu():
    # Format exception, invalid operation exception
    try:
        # Initialise integer variable
        choice = -1
        # while loop for continuing with menu or not
        while choice != 0:
            # Display main menu
            print("***************************************************************************************************")
            print("                             LANGHAM HOTEL MANAGEMENT SYSTEM                                       ")
            print("                                         MENU                                                      ")
            print("***************************************************************************************************")
            print(" 1. Add Room")
            print(" 2. Delete Room")
            print(" 3. Display Rooms Details")
            print(" 4. Allocate Rooms")
            print(" 5. Display Room Allocation details")
            print(" 6. Billing & De-Allocation Details")
            print(" 7. Save the Room Allocations To a File")
            print(" 8. Show the room Allocations From a File")
            print(" 9. BackUp")
            print(" 0. Exit Application")
            print("***************************************************************************************************")
            print("Please Enter Your Choice Number Here: ")

            # Ask user to enter a number for menu choice
            choice = int(input("Enter Your Choice Number Here (0-9): "))

            if choice == 0:
                # Exit App
                exit(0)

            elif choice == 1:
                # Add Rooms
                add_room()

            elif choice == 2:
                # Delete Room
                delete_room()

            elif choice == 3:
                # Allocate Room
                display_rooms_details()

            elif choice == 4:
                # De-Allocate Room
                allocate_rooms()

            elif choice == 5:
                # Display Room Details
                display_room_allocations_details()

            elif choice == 6:
                # Billing & De-Allocation Details
                billing_and_de_allocation_details()

            elif choice == 7:
                # Save the Room Allocations To a File
                save_the_room_allocations_to_a_file()

            elif choice == 8:
                # Show the room Allocations From a File
                show_the_room_allocations_from_a_file()

            elif choice == 9:
                # BackUp
                backup()

                # Invalid Inputs
            else:
                print("Please Enter a Valid Number Between 0-9")

    except SyntaxError as e:
        print(f"Syntax Error : {e} ")

    except ValueError as e:
        print(f"Value Error : {e} ")

def add_room():
    global no0fRoom, list0fRooms
    # Format exception
    try:
        # Display 'add rooms' choice selected
        print("You have selected 'ADD ROOMS' from menu")
        # Ask user how many rooms to add, and convert and save as integer variable
        no0fRoom = int(input("Please enter the total number of rooms in the hotel"))
        # Display how many rooms are there in total
        print(f"Hotel has {no0fRoom} rooms in total")
        print ("******************************************************************************************************")
        # To check list (list0fRoom) is not empty
        if len(list0fRooms) > 0:
            y = len(list0fRooms)
        else:
            y = 0
        # Create new array with number of rooms to add
        for x in range(no0fRoom):
            room = Room()
            list0fRooms.append(room)
        # Use for loop to ask user to input room details and save it
        for i in range(y, y + no0fRoom):
            obj_room = list0fRooms[i]
            print(f"Room Allocation {i+1}:")
            # Ask user to enter room number
            obj_room.RoomNo = int(input(f"Please enter the room number {i+1} :"))
            # Declare room allocation status to false (not allocated)
            obj_room.IsAllocated = False
            # Save the
            list0fRooms[i] = obj_room
            # If condition to check the same room number exist or not when there is more than 1 room added

            if i > 0:
                # Use for loop to check the list0fRooms with the same room number
                for j in range(i):
                    # Use while loop to check the user input room number is equal to
                    while list0fRooms[i].RoomNo == list0fRooms[j].RoomNo:
                        # Display message for the same room exist and ask to input new room n
                        print(f"Same room number already exist")
                        obj_room.RoomNo = int(input(f"Please enter a new room number {i+1}:"))
                        # Declare room allocation status to false (not allocated)
                        obj_room.IsAllocated = False
                        list0fRooms[i] = obj_room

    except ValueError as e:
        print(f"Value error : {e} ")
        print("Invalid input, Please try again ")
        add_room()



def delete_room():
    global list0fRooms
    # If condition to check there is any room to delete
    if list0fRooms:
        # Display 'delete rooms' choice selected
        print("You have selected 'DELETE ROOM' from menu")
        print("*******************************************************************************************************")
        room_Number = int(input("Please enter the room number that you want to delete "))
        try:
            for obj_room in list0fRooms:
                if obj_room.RoomNo == room_Number:
                    list0fRooms.remove(obj_room)
        except ValueError as e:
            print(f"Value error : {e} ")
            print("Invalid input, Please try again ")
            delete_room()
    # If nothing to delete, display following message
    else:
        print("No rooms to delete\nPlease add rooms first")

def display_rooms_details():
    global list0fRooms
    # If condition to check there is any rooms to display
    if list0fRooms:
        # Display 'display room' choice selected
        print("You have selected have been added")
        print("*******************************************************************************************************")
        print(f"Following rooms have been added")
        for obj_room in list0fRooms:
            print(f"Room Number : {obj_room.RoomNo}")
    # If nothing to display, display following message
    else:
        print("No rooms to display\nPlease add rooms first")


def allocate_rooms():
    global no0fRooms, list0fRooms, list0fRoomAllocations
    try:
        print("You have selected 'ALLOCATE ROOMS' from menu")
        allocate_room = int(input("How many rooms would you like to allocate?: "))
        while allocate_room > len(list0fRooms):
            # Display input is incorrect message and give another chance
            print(f"You cannot allocate more rooms than the total number of rooms in the Hotel\n"
                  f"Please enter a number between 1-{len(list0fRooms)}: ")
            allocate_room = int(input("How many rooms would you like to allocate?: "))
        print(f"You are allocating {allocate_room} room(s)")
        i = 0
        while i < allocate_room:
            print("***************************************************************************************************")
            roomAllocation = RoomAllocation()
            customer = Customer()
            print(f"Room Allocation {i + 1}:")
            searchRoom = int(input("Please search Room Number to allocate: "))
            for j in range(len(list0fRooms)):
                if searchRoom == list0fRooms[j].RoomNo:
                    print("Found matching room number to allocate")
                    if not list0fRooms[j].IsAllocated:
                        print(f"Room {list0fRooms[j].RoomNo} is empty ")
                        customer.CustomerNo = int(input("Please enter Customer Number to allocate: "))
                        customer.CustomerName = int(input("Please enter Customer Name to aloocate: "))
                        list0fRooms[j].IsAllocated = True
                        print("Allocation has been done")
                        roomAllocation.AllocatedRoomNo = searchRoom
                        roomAllocation.AllocatedCustomer = customer
                        list0fRoomAllocations.append(roomAllocation)

                        break;
                    else:
                        print(f"Room {list0fRooms[j].RoomNo} is already occupied\n"
                              f" Please enter another room to allocate")
                        i -= 1
                        break;
                else:
                    while j == len(list0fRooms) -1:
                        print(f"Could not find matching room number\n" +
                              f"Please enter correct room number or add room first")
                        i -= 1
                        break

            i += 1
    except ValueError:
        print("Invalid input. Please try again.")
        allocate_rooms()



def display_room_allocations_details():
    global list0fRoomAllocations
    # If condition to check there is any rooms to display
    if list0fRoomAllocations:
        # Display 'display room' choice selected
        print("You have selected 'DISPLAY ROOMS ALLOCATION DETAILS' from menu")
        print("*******************************************************************************************************")
        print(f"Following rooms have been allocated")
        for obj_room in list0fRoomAllocations:
            print(f"Allocated Room Number : {obj_room.AllocatedRoomNo}")
            print(f"Customer Number : {obj_room.AllocatedCustomer.CustomerNo}")

    # If nothing to display, display following message
    else:
        print("No allocated rooms to display\nPlease add rooms first")
def billing_and_de_allocation_details():
    global no0fRooms, list0fRooms, list0fRoomAllocations
    try:
        print("You have selected 'BILLING & DE-ALLOCATION' from menu")
        print("*******************************************************************************************************")
        

def save_the_room_allocations_to_a_file():
    global filePath
    try:
        print("You have selected 'SAVE THE ROOM ALLOCATIONS TO A FILE' from menu")
        print("*******************************************************************************************************")
        with open(filePath, "w") as file:
            now = str(datetime.now())
            for obj_roomAllocated in list0fRoomAllocations:
                add_to_file = f"*********************************************************************************\n" \
                              f"Room Number: {obj_roomAllocated.AllocatedRoomNo}\n" \
                              "Customer Number: {obj_roomAllocated.AllocatedCustomer.CustomerNo}\n" \
                              f"Customer Name: {obj_roomAllocated.AllocatedCustomer.CustomerName}\n" \
                              f"Current date and time is {now}"
                file.wirte(add_to_file + "\n")

        print("File saved as 'lhms_764702333.txt' under Documents folder")
    except IOError as ex:
        print(ex)
        print("Please try again")
    except FileNotFoundError as ex:
        print(ex)
        print("Please try again")

def show_the_room_allocations_from_a_file():
    global filePath
    try:
        print("You have selected 'SHOW THE ROOM ALLOCATIONS FROM A FILE' from menu")
        print("*******************************************************************************************************")
        with open(filePath, "r") as file:
            lines = file.readlines()
            for line in lines:
                print(line.strip())

    except IOError as ex:
        print(ex)
        print("Please try again")
    except FileNotFoundError as ex:
        print(ex)
        print("Please try again")
    except EOFError as ex:
        print(ex)
        print("Please try again")

def backup():
    global filePath, filePathBackUp
    print("You have selected 'BACKUP' from menu")
    print("*******************************************************************************************************")
    while os.path.exists(filePathBackUp):
        print("'lhms_764702333' file already exist\nExisting file will be deleted")
        os.remove(filePathBackUp)

    os.rename(filePath, filePathBackUp)
    print()
if __name__ == '__main__':
    main()





```
